### Data Visualization

- Visualizing the data in our dataset

```python
#Creating a new dataframe

df = df_covid[["new_cases","new_deaths","population","aged_65_older_percent","gdp_per_capita","hospital_beds_per_thousand"]]
df
```

```python
df.corr()
```

```python
#Checking the correlation  of the columns

corr = df.corr(method = "spearman")
sns.heatmap(corr,annot = True)
plt.show()
```
