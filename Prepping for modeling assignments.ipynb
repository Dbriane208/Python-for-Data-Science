```python
# Importing the required libray
import pandas as pd
```

```python
# Read all four files into a Jupyter Notebook
April_sales = pd.read_excel('Data/Book_Sales_April.xlsx')
May_sales = pd.read_excel('Data/Book_Sales_April.xlsx')
June_sales = pd.read_excel('Data/Book_Sales_April.xlsx')
customers = pd.read_csv('Data/Customers.csv')

print('Loading data done')
```

```python
# viewing april sales table
April_sales.head()
```

```python
# viewing may sales table
May_sales.head()
```

```python
# viewing june sales
June_sales.head()
```

```python
# viewing the customers data
customers.head()
```

```python
# Renaming the customer id column
customers.rename(columns = {'CustomerID':'Customer ID'},inplace=True)
customers.head()
```

```python
# Concatenate the April, May and June book sales DataFrames
sales = pd.concat([April_sales,May_sales,June_sales]).reset_index(drop=True)
sales.head()
```

```python
# Explore the data
sales.shape
```

```python
# Join the newly created book sales DataFrame with the customers DataFrame on customer_id
# Which type of join do you think would work best here? - left join because we want details of all the customers
df = sales.merge(customers,how='left',on='Customer ID')
df.head(10)
```
