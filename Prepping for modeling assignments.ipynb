```python
# Importing the required libray
import pandas as pd
```

```python
# Read all four files into a Jupyter Notebook
April_sales = pd.read_excel('Data/Book_Sales_April.xlsx')
May_sales = pd.read_excel('Data/Book_Sales_May.xlsx')
Junesales = pd.read_excel('Data/Book_Sales_June.xlsx')
customers = pd.read_csv('Data/Customers.csv')

print('Loading data done')
```

```python
# Renaming the customer id column
customers.rename(columns = {'CustomerID':'Customer ID'},inplace=True)
customers.head()
```

```python
# Concatenate the April, May and June book sales DataFrames
sales = pd.concat([April_sales,May_sales,Junesales]).reset_index(drop=True)
sales.head()
```

```python
# Explore the data
sales.shape
```

```python
# Join the newly created book sales DataFrame with the customers DataFrame on customer_id
# Which type of join do you think would work best here? - left join because we want details of all the customers
df = sales.merge(customers,how='left',on='Customer ID')
df.sample(10)
```

## 2. Prepare Rows for Modeling

Hi again,

We are trying to predict which customers will purchase a book this month.

Can you reformat the data you compiled earlier this week so that it’s ready to be input into a model, with each row representing a customer instead of a purchase?

Thanks!\
Brooke

```python
# What should one row of data look like for the model?
# must have a customer colum - given by purchased on june or not and a total spend column -
# given by total spend on may and april
```

```python
# Create a column called 'June Purchases' that sums up all purchases in June
juneRows = df[df['Purchase Date'].dt.month == 6]
juneRows.sample(5)
```

```python
junePurchase = juneRows.groupby('Customer ID')['Price'].count().rename('June Purchases').to_frame().reset_index()
junePurchase
```

```python
# Create a column called 'Total Spend' that sums up the prices of all the books that were purchased in April and May
aprilMay = df[df['Purchase Date'].dt.month < 6]
aprilMay.sample(5)
```

```python
total_spend = aprilMay.groupby('Customer ID')['Price'].sum().rename('total_spend').to_frame().reset_index()
total_spend
```

```python
# Combine the 'June Purchases' and 'Total Spend' columns into a single DataFrame for modeling
# Concat does not preserve the customer id column it creates another since it stacks
data_concat = pd.concat([junePurchase,total_spend],axis=1)
data_concat
```

```python
# Using merge to combine the data frames
df_data = junePurchase.merge(total_spend)
df_data
```

## 3. Prepare Columns for Modeling

```python
# Change the Audience field into a dummy variable
aud_dummies = pd.get_dummies(aprilMay.Audience)
aud_dummies.head()
```

```python
# Using the Audience dummy variables, create three new columns that contain
# the number of Adult / Children / Teen books purchased by each customer
data = pd.concat([aprilMay['Customer ID'],aud_dummies],axis=1)
data.head()
```

```python
# Grouping the data
book_data = data.groupby('Customer ID').sum()
book_data
```

```python
# Joining with the customer level data
custData = df_data.merge(book_data,how='left',on='Customer ID')
custData
```
