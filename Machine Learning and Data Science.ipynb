```python
#According to the pie chart I wouls like to gain the insights between newcases and the new deaths, population,gdp_per_capita
#aged_65_older_percent, hospital_beds_per_thousand

sns.relplot(x = "new_cases",y = "new_deaths" ,data = df, kind = "scatter")
plt.show()
```

```python
#We can deduce that as the population increases the number of cases tend to increase

sns.relplot(x = "population",y = "new_cases" ,data = df, kind = "line")
plt.show()
```

```python
# In places we have low number of people we have many people who are over 65 percent
# The virus is more active on areas we have old population than in young population

sns.relplot(x = "population",y = "aged_65_older_percent" ,data = df, kind = "line")
plt.show()
```

```python
# We can see that in places that have high gdp per capita the number of covid cases are low and vice versa
# Developed countries have effective ways to curb the covid virus

sns.relplot(x = "new_cases",y = "gdp_per_capita" ,data = df, kind = "line")
plt.show()
```

```python
# In countries that had high gdp per capita the number of deaths were less 

sns.relplot(x = "gdp_per_capita",y = "new_deaths" ,data = df, kind = "scatter")
plt.show()
```

```python
# We identify that there are strong relationship between the number of deaths, hospital beds and the gdp of a country

sns.relplot(x = "hospital_beds_per_thousand",y = "new_deaths" ,data = df, kind = "scatter")
plt.show()
```

## Creating and training a model

- We create and train a model to predict the outcomes given certain attributes

```python
df.head(3)
```

```python
# Let's have our target(y) and the attributes(x)

x = df.drop("new_deaths", axis = 1)
y = df["new_deaths"]

print("Split done")
```

```python
# We split our data into test and train datasets

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2)

print("Split done")
```

```python
# We train the model through different models

#from sklearn.tree import DecisionTreeRegressor
#model =  DecisionTreeRegressor()
#model.fit(x_train,y_train)


#print(model.score(x_test,y_test))
```

```python
# From the model tested we can the use it to predict other features

from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
model.fit(x_train,y_train)

print(model.score(x_test,y_test))
```

```python
# Lets predict new_deaths using new data
# The procedure is new_cases	population	aged_65_older_percent	gdp_per_capita	hospital_beds_per_thousand

features = np.array([[100,45000000,16,2500,10]])
print(model.predict(features))

#Given the above dataset our model predicts that there will be 8.8 deaths
```
