## Filtering

```python
# Importing the packages
import pandas as pd
```

```python
# Loading the dataset
df = pd.read_pickle("Data/groceries_with_new_columns.pkl")
df.head(5)
```

```python
# Cleaning the Subcategory column
df.Subcategory = df.Subcategory.str.strip()
df.Subcategory[0]
```

```python
# Using the loc accessor to filter
df.loc[df.Subcategory == 'Vegetable',['Subcategory','Item','Price_Dollars']]
```

```python
# Filtering without loc accessor
df[df.Item == 'Apple']
```

```python
# Using the mask boolean to do multiple filter
mask = (df.Category == 'Produce') & (df.Price_Dollars >= 5)
df.loc[mask,['Category','Item','Price_Dollars']]
```

## Sorting

```python
# Sorting from the lowest to the highest
df.sort_values('Price_Dollars').head(5)
```

```python
# Sorting from the highest to the highest
df.sort_values('Inventory',ascending=False).tail(5)
```

```python
# Chaining the sorting and filtering
mask = (df.Category == 'Produce') & (df.Subcategory == 'Fruit')
df.loc[mask,['Category','Subcategory','Item','Inventory']].sort_values('Inventory',ascending=False)
```
