## Converting Data types

```python
# Importing Pandas
import pandas as pd
```

```python
# Loading the dataset
data_runtime = pd.read_excel("Data/Run Times.xlsx")
data_runtime
```

```python
#  Checking the datatypes of our dataset
data_runtime.info()
```

```python
data_runtime.dtypes
```

```python
# Converting the Fee column to a float64 datatype
data_runtime["Fee"] = data_runtime.Fee.str.replace('$','')
data_runtime
```

```python
data_runtime.dtypes
```

```python
# After removing the dollar sign we convert the column to float64 datatype
data_runtime.Fee = pd.to_numeric(data_runtime["Fee"])
data_runtime
```

```python
# Now if we run the cell the datatype will be of type float
data_runtime.dtypes
```

```python
# Converting the Warm up time to a numeric dtype
data_runtime["Warm Up Time"] = data_runtime["Warm Up Time"].astype("str").str.replace(" min","")
data_runtime
```

```python
data_runtime.dtypes
```

```python
# Converting the Warm Up Time to a numeric data type
data_runtime["Warm Up Time"] = pd.to_numeric(data_runtime["Warm Up Time"])
data_runtime
```

```python
data_runtime.dtypes
```

```python
# Changing the Rain Coloumn to a numeric type
data_runtime.Rain = data_runtime.Rain.astype("int")
data_runtime
```

## Findig Missing Data

```python
# Loading the data
students = pd.read_excel("Data/Student Grades.xlsx")
students
```

```python
# Checking the null values
students.Grade.value_counts(dropna=False)
```

```python
students.isna()
```

```python
# Checking the number of missing data in every column
students.isna().sum()
```

```python
# Checking null values using a boolean
# Missing value -> True
# No missing value -> False
students.isna().any(axis=1)
```

## Handling Missing Data

```python
# Rows that has missing values
students[students.isna().any(axis=1)]
```

```python
# Dropping some rows with Student and class missing data
students[students.isna().any(axis=1)].dropna(subset=['Student','Class'])
```

```python
# If we run these command it will erase all the rows with missing data
students[students.isna().any(axis=1)].dropna()
```

```python
students[students.isna().any(axis=1)]
```

```python
# Even after we dropped the rows with missing data in class and student our data didn't get updated
# To update our data we need to set the new data back to our dataframe
students = students.dropna(subset=['Student','Class'])
students
```

```python
# Checking whether we still have the data we dropped
students[students.isna().any(axis=1)]
```

```python
# We can use inplace = true to update our dataset everytime a change is made
students.dropna(subset=['Student','Class'],inplace=True)
students
```

```python
# We can confirm by checking our data shape
students.shape
```

```python
# Loading in the dataset
df = pd.read_excel("Data/Alarm Survey Data.xlsx")
df
```

```python
# Checking for any missing data
df.isna().sum()
```

```python
# Checking which data is missing
df.info()
```

```python
# Checking the null values
df[df.isna().any(axis=1)]
```

```python
# Checking for unique values in the sleep_quality column
# Use argument dropna=False to check which value is NaN
df["sleep_quality"].value_counts(dropna=False)
```

```python
# From the look of value_count progression the NaN values should be a 2
df["sleep_quality"].fillna(2).value_counts(dropna=False)
```

```python
# Checking whether the data is updated
df.info()
```

```python
# Using the inplace=True argument
df["sleep_quality"].fillna(2,inplace=True).value_counts(dropna=False)
```
