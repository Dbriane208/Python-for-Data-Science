## Data Manipulation using pandas

```python
#import the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the data

df = pd.read_csv("university_enrollment_2306.csv")
df
```

```python
#Getting a random sample of our data
df.sample(10)
```

```python
#Getting the shape of our data
df.shape
```

```python
#Checking the missing values
df.isna().sum()
```

```python
#Dropping the pre_requirement
df = df.drop(["pre_requirement"],axis = 1)
df
```

```python
#Filling the missing values of post_score with the mean
df.fillna(np.mean(df["post_score"]),inplace = True)
```

```python
#Checking the missing values after filling post score
df.isna().sum()
```

```python
#Checking the value counts of course_type
df["course_type"].value_counts()
```

```python
#Checking the value counts of year
df["year"].value_counts()
```

```python
#Checking the sum of course_type column
df["course_type"].sum
```

```python
#Checking the sum of course_type column
df["year"].sum
```

```python
#Checking the value counts of department
df["department"].value_counts()
```

```python
#Changing the the categorical values to numeric values
new_df = pd.get_dummies(df,columns = ["department"])
new_df
```

```python
#Getting the statistics of the data
new_df.describe()
```

```python
#Getting the new information of our dataset
new_df.info()
```

```python
#Checking the correlation of the data
new_df.corr()
```

```python
#Drawing the heatmap
corr = new_df.corr(method = "spearman")
sns.heatmap(corr,annot = True)
plt.show()
```

```python
x = new_df.drop("course_type", axis = 1)
y = new_df["course_type"]

print("Split done")
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2)

print("Split done")
```

```python
from sklearn.tree import DecisionTreeRegressor

model =  DecisionTreeRegressor()
model.fit(x_train,y_train)


print(model.score(x_test,y_test))
```
