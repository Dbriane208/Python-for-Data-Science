## Creating and Training a model

```python
#Spliting data int x alabels and y target

x = df.drop(["smoke"],axis = 1 )
y = df["smoke"]

print("Spliting done")
```

```python
#Importing and trianing the model

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)

print("Training Done")
```

```python

# Finding the optimum number of clusters for k-means classification

x = df.iloc[:, [0, 1, 2, 3]].values

from sklearn.cluster import KMeans
wcss = [] #wcss-the sum of squared distance between each point and the centroid in a cluster.

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

```python
# Importing model

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)

print("Fitting Done")
```

```python
#Checking the score of the model

ml_model = model.score(x_test,y_test)
ml_model
```

```python
#Let's predict tthe smoke output given the input attributes
features = np.array([[5,10000,15789,2,200,90,157,100,3]])
print(model.predict(features))
```

```python
#Save the model
import pickle 
pickle.dump(model,open("Logistic_regression_model.pk1","wb"))

loaded_model = pickle.load(open("Logistic_regression_model.pk1","rb"))
loaded_model.score(x_test,y_test)
```
